
# Discussion

This study aimed to answer: **How do room characteristics, location features, and customer ratings affect Airbnb listing prices in major European cities?**

## Summary of Findings with Respect to the Goal
- **Room Characteristics:** Larger properties, more bedrooms, and higher amenity counts are associated with higher prices, though with diminishing returns for size.
- **Location Features:** Listings closer to the city center and metro stations generally have higher prices, but the relationship flattens at very short distances, indicating saturation.
- **Customer Ratings:** Higher guest satisfaction and cleanliness ratings are linked to increased prices, especially in premium listings. Median prices are stable across most cleanliness levels, but high-end prices are more common with top ratings.
- **Modeling Outcome:** ANOVA tests confirmed that a **full quadratic OLS regression model** (including squared terms for cleanliness, satisfaction, and distances) significantly improves fit over a reduced model. This model was selected as the final predictive model; LASSO regularization did not improve RMSE.

## Were These Results Expected?
Yes, the results were broadly in line with both market expectations and the assumptions of our chosen model:

1. **Linearity in Parameters:**  
   We expected that the log(price) could be modeled as a linear combination of parameters, and including polynomial terms allowed us to capture important nonlinear effects without violating this assumption. The observed curvature in relationships (e.g., diminishing returns for distance and size) confirmed the usefulness of these terms.

2. **Independence:**  
   While we assumed independence of observations, we anticipated that some mild dependence might exist (e.g., multiple listings by the same host). The similarity of results across cities suggests any such clustering did not dominate the findings.

3. **Homoscedasticity:**  
   The log transformation of price was expected to stabilize variance, and residual plots showed relatively constant spread, supporting this expectation.

4. **Normality of Residuals:**  
   Given the large sample size, we expected minor deviations from normality to be unimportant for inference. Residual diagnostics showed approximate normality, validating this assumption.

5. **No Perfect Multicollinearity:**  
   Based on exploratory checks, we expected low to moderate correlation between predictors. VIF results confirmed this, allowing reliable coefficient estimation.
